```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io



file_name = next(iter(uploaded))  # Get the filename from the uploaded dictionary
df = pd.read_csv(io.BytesIO(uploaded[file_name])) # Read the file into a pandas DataFrame

# Now you can proceed with your operations on the DataFrame
# Drop irrelevant columns (e.g., 'Unnamed: 0' if itâ€™s just an index)
df.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore') # Added errors='ignore' to handle cases where 'Unnamed: 0' might not exist

# Check for missing values
print(df.isnull().sum())
```

```python
def get_weekend_destinations(city_name, df, max_travel_time=6, top_n=5):
    # Get the state of the input city
    city_row = df[df['City'].str.lower() == city_name.lower()]

    if city_row.empty:
        return f"No data available for the city: {city_name}"

    state = city_row['State'].values[0]

    # Filter places within the same state
    nearby_places = df[df['State'] == state]

    # Filter places that can be visited within a reasonable time (e.g., 2-6 hours)
    nearby_places = nearby_places[
        (nearby_places['time needed to visit in hrs'] >= 2) &
        (nearby_places['time needed to visit in hrs'] <= max_travel_time)
    ]

    # Sort by Google review rating and number of reviews
    ranked_places = nearby_places.sort_values(
        by=['Google review rating', 'Number of google review in lakhs'],
        ascending=[False, False]
    )

    # Return top N places
    return ranked_places[['Name', 'City', 'Google review rating', 'Number of google review in lakhs']].head(top_n)

# Example usage
city_name = "Delhi"
recommended_places = get_weekend_destinations(city_name, df)
print(recommended_places)
```

```python
# Get user input
user_city = input("Enter your city: ")

# Get travel recommendations
recommendations = get_weekend_destinations(user_city, df)
print(recommendations)
```
